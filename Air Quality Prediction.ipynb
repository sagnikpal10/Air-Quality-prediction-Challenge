```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Loading the Dataset

```python
df = pd.read_csv('Train.csv')
print(df.columns)
```

```python
y = list(df.get('target'))
print(y)
```

```python
df.values
```

```python
x = df.values[:,0:5]
print(x)
y = df.values[:,5]
print(y)
```

```python
print(x.shape)
print(y.shape)
```

## Training the model

```python
from sklearn.linear_model import LinearRegression
```

```python
# 1. Creating an object
model = LinearRegression(normalize=True)
```

```python
# 2. Training the model
model.fit(x,y)
```

```python
# 3. Printing the coefficients
print(model.coef_)
print(model.intercept_)
```

## Acuracy of Prediction

```python
## Accuracy of prediction on the training data
model.score(x,y)
```

```python
## Loading the test data
dftest = pd.read_csv('Test.csv')
x_test = dftest.values
```

```python
y_test = model.predict(x_test)
print(type(y_test))
# print(x_test)
# print(y_test)
## Accuracy prediction of the test data
print(model.score(x_test,y_test))
```
